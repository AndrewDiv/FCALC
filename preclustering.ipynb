# Iris

```python
iris_df = pd.read_csv('data_sets/iris.data',header=None, names=['sepal length', 'sepal width', 'petal length', 'petal width', 'species'])
iris_df.head()
```

```python
fig, axes = plt.subplots(1,4, figsize=(20, 5))

fig.suptitle('Numeric features disribution in Iris dataset', fontsize=20)

sns.histplot(ax=axes[0], data=iris_df, x='sepal length', hue='species', multiple='stack'); axes[0].set(ylabel=None)
sns.histplot(ax=axes[1], data=iris_df, x='sepal width', hue='species', multiple='stack'); axes[1].set(ylabel=None)
sns.histplot(ax=axes[2], data=iris_df, x='petal length', hue='species', multiple='stack'); axes[2].set(ylabel=None)
sns.histplot(ax=axes[3], data=iris_df, x='petal width', hue='species', multiple='stack'); axes[3].set(ylabel=None)

plt.show()
```

```python
fig, axes = plt.subplots(1,4, figsize=(20, 5))

fig.suptitle('Numeric features disribution in Iris-setosa species', fontsize=20)

sns.histplot(ax=axes[0], data=iris_df.loc[iris_df['species']=='Iris-setosa'], x='sepal length', kde=True); axes[0].set(ylabel=None)
sns.histplot(ax=axes[1], data=iris_df.loc[iris_df['species']=='Iris-setosa'], x='sepal width', kde=True); axes[1].set(ylabel=None)
sns.histplot(ax=axes[2], data=iris_df.loc[iris_df['species']=='Iris-setosa'], x='petal length', kde=True); axes[2].set(ylabel=None)
sns.histplot(ax=axes[3], data=iris_df.loc[iris_df['species']=='Iris-setosa'], x='petal width', kde=True); axes[3].set(ylabel=None)
plt.show()
```

```python
fig, axes = plt.subplots(1,4, figsize=(20, 5))

fig.suptitle('Numeric features disribution in Iris-versicolor species', fontsize=20)

sns.histplot(ax=axes[0], data=iris_df.loc[iris_df['species']=='Iris-versicolor'], x='sepal length', kde=True, color=sns.color_palette("tab10")[1]); axes[0].set(ylabel=None)
sns.histplot(ax=axes[1], data=iris_df.loc[iris_df['species']=='Iris-versicolor'], x='sepal width', kde=True, color=sns.color_palette("tab10")[1]); axes[1].set(ylabel=None)
sns.histplot(ax=axes[2], data=iris_df.loc[iris_df['species']=='Iris-versicolor'], x='petal length', kde=True, color=sns.color_palette("tab10")[1]); axes[2].set(ylabel=None)
sns.histplot(ax=axes[3], data=iris_df.loc[iris_df['species']=='Iris-versicolor'], x='petal width', kde=True, color=sns.color_palette("tab10")[1]); axes[3].set(ylabel=None)
plt.show()
```
